```python
t,s=var('t,s')
f(t,s)=1-((s*cos(4*t)*cos(t))/(sqrt(1-s^2*(cos(4*t)^2)*(sin(t)^2))))
s = 0.5
x(t)=f(t-pi/2, s)
y(t)=f(t, s)
parametric_plot((x(t),y(t)), (t,0,2*pi))
```

```python
t,s=var('t,s')
f(t,s)=1-(s*cos(4*t)*cos(t))/(sqrt(1-s^2*(cos(4*t))^2*(sin(t))^2))
for k in [1..10]:
    k = k/10
    x(t)=f(t-pi/2, k)
    y(t)=f(t, k)
    g1=parametric_plot((x(t),y(t)), (t,0,2*pi),color='cyan')
    g=g+g1
show(g)
```
